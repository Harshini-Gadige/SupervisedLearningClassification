```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = "/Users/harshini/Desktop/Thinkful/sentiment labelled sentences/imdb.txt"
df = pd.read_csv(filepath, sep='\t',names = ["Sentence", "Score"])
```

```python
df.head()
```

```python
keys = ['good','excellent','great','impressed','recommend','best','sweet','fine','well','nice']
for key in keys:
    df[str(key)] = df.Sentence.str.contains(str(key),case=False)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(), square=True, annot=True, linewidths=.5)
plt.show()
```

```python
data = df[keys]
target = df['Score']
```

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
model.fit(data,target)
y_pred = model.predict(data)
print("Number of mismatched values out of total {} are :  {}".format(data.shape[0],(y_pred!=target).sum()))
```

```python
# Amazon : Number of mismatched values out of total 1000 are :  281
# Yelp : Number of mismatched values out of total 1000 are :  348
# imdb : Number of mismatched values out of total 748 are :  311
```
